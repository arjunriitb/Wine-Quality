```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error
```

```python
# Split the red wine data into features (X_red) and target variable (y_red)
X_red = red_wine_df.drop(columns=['quality'])  # Features
y_red = red_wine_df['quality']  # Target variable

# Split the white wine data into features (X_white) and target variable (y_white)
X_white = white_wine_df.drop(columns=['quality'])  # Features
y_white = white_wine_df['quality']  # Target variable

# Split the red wine data into training (70%), validation (15%), and test (15%) sets
X_red_train, X_red_temp, y_red_train, y_red_temp = train_test_split(X_red, y_red, test_size=0.3, random_state=42)
X_red_val, X_red_test, y_red_val, y_red_test = train_test_split(X_red_temp, y_red_temp, test_size=0.5, random_state=42)

# Split the white wine data into training (70%), validation (15%), and test (15%) sets
X_white_train, X_white_temp, y_white_train, y_white_temp = train_test_split(X_white, y_white, test_size=0.3, random_state=42)
X_white_val, X_white_test, y_white_val, y_white_test = train_test_split(X_white_temp, y_white_temp, test_size=0.5, random_state=42)
```

This train_test_split function is used to split the data into training and temp varibles. Further temp variables are split into validation and testing data. Hence train variables for training the model, validation varibles for finding the hyperparameters and finally the test varibles for testing the model.

```python
# RED WINE
# Train Random Forest model
rf_model_red = RandomForestRegressor(random_state=42)
rf_param_grid = {'n_estimators': [100, 200, 300],
                 'max_depth': [None, 10, 20]}
rf_grid_search = GridSearchCV(rf_model_red, rf_param_grid, cv=3, scoring='neg_mean_squared_error')
rf_grid_search.fit(X_red_train, y_red_train)
best_rf_model_red = rf_grid_search.best_estimator_
```

Here using Random forest Regressor model is trained. The hyperparameters being tuned are
 n_estimators: defines the number of decision trees to be created in the forest.
 max_depth : gives the maximum depth to be given, if it is 'None' then nodes are expanded until all leaves are pure or until all leaves contain less than min_samples_split samples. I have used chatgpt to implement this code and understanding about the hyperparameters used.

```python
# Validate Random Forest model
rf_val_predictions_red = best_rf_model_red.predict(X_red_val)
rf_val_mse_red = mean_squared_error(y_red_val, rf_val_predictions_red)
print("Random Forest Validation MSE for red wine:", rf_val_mse_red)
```

By validating using validation set optimal hyperparametrs for Random Forest are found. The GridSearchCV function from scikit-learn iterates over all possible combinations of these hyperparameters.

```python
# Train SVR model
svr_model_red = SVR(kernel='rbf')
svr_param_grid = {'C': [0.1, 1, 10],
                  'gamma': ['scale', 'auto']}
svr_grid_search = GridSearchCV(svr_model_red, svr_param_grid, cv=3, scoring='neg_mean_squared_error')
svr_grid_search.fit(X_red_train, y_red_train)
best_svr_model_red = svr_grid_search.best_estimator_

# Validate SVR model
svr_val_predictions_red = best_svr_model_red.predict(X_red_val)
svr_val_mse_red = mean_squared_error(y_red_val, svr_val_predictions_red)
print("SVR Validation MSE for red wine:", svr_val_mse_red)
```

For SVR model the hyperparametrs are C and gamma. C controls the trade-off between a smooth decision boundary and classifying the training points correctly. A smaller value of C leads to a smoother decision boundary, while a larger value of C leads to a decision boundary that correctly classifies more training points. Gamma  defines how much influence a single training example has. A smaller value of gamma means a larger radius of influence for each data point, resulting in a smoother decision boundary. while, a larger value of gamma makes the decision boundary more localized around each data point.
I have used ChatGPT for implementing this code and understanding about the hyperparameters used.

```python
# Test the best-performing models (Random Forest and SVR) using the test set
rf_test_predictions_red = best_rf_model_red.predict(X_red_test)
rf_test_mse_red = mean_squared_error(y_red_test, rf_test_predictions_red)
print("Random Forest Test MSE for red wine:", rf_test_mse_red)

svr_test_predictions_red = best_svr_model_red.predict(X_red_test)
svr_test_mse_red = mean_squared_error(y_red_test, svr_test_predictions_red)
print("SVR Test MSE for red wine:", svr_test_mse_red)
```

From the MSE values obtained for the test sets of Random Forest and SVR it is evident that Random Forest is a better model for the given problem (Random Forest has least value of MSE)

```python
# Get feature importances for Random Forest model
feature_importances_rf = best_rf_model_red.feature_importances_

# Map feature importances to feature names
feature_names = X_red_train.columns
importance_dict_rf = dict(zip(feature_names, feature_importances_rf))

# Print feature importance scores
for feature, importance in importance_dict_rf.items():
    print(f"{feature}: {importance}")
```

The importance of features for the Random Forest model is found by finding the decrease in impurity by each feature.Higher the decrease in impurity more the importance of that feature.
I have understood the concept through this video https://www.youtube.com/watch?v=cw4qG9ePZ9Y
And I have used chatgpt to implement the code.  


```python
# Get support vectors
support_vectors = best_svr_model_red.support_vectors_

# Get dual coefficients (weights) of support vectors
dual_coef = best_svr_model_red.dual_coef_[0]

# Calculate feature importance based on support vectors and dual coefficients
feature_importance_svr = np.abs(np.dot(dual_coef, support_vectors))

# Map feature importances to feature names
importance_dict_svr = dict(zip(feature_names, feature_importance_svr))

# Print feature importance scores
for feature, importance in importance_dict_svr.items():
    print(f"{feature}: {importance}")
```

In SVR with an RBF kernel, feature importance is indirectly estimated by analyzing the support vectors and their associated dual coefficients. The contribution of each feature to the decision boundary is calculated by taking the dot product of the dual coefficients and the corresponding support vectors for each feature. Positive contributions indicate features important for predicting higher target values, while negative contributions indicate features important for predicting lower target values. Taking the absolute values of these contributions provides an overall measure of feature importance, offering insights into the relative importance of features in the SVR model's decision-making process.

The features 'Alcohol' , 'volatile Acidity' , 'Sulphates' and ' Total Sulfur dioxide' are the ones which are given importance by both the models, rest of the features' importance order are not the same.
